```python
import pandas as pd
import matplotlib.pyplot as plt

cars = pd.read_csv(r"C:\Users\prane\Downloads\Cars.csv")

plt.figure(figsize=(12, 6))
plt.bar(cars.index, cars['HP'], label='HP', alpha=0.7, color='black')
plt.bar(cars.index, cars['MPG'], label='MPG', alpha=0.7, color='g')

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Bar Plot of HP and MPG")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

cars = pd.read_csv(r"C:\Users\prane\Downloads\Cars.csv")

plt.figure(figsize=(12, 6))
plt.bar(cars.index, cars['HP'], label='HP', alpha=0.7, color='black')
plt.bar(cars.index, cars['VOL'], label='VOL', alpha=0.7, color='g')

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Bar Plot of HP and MPG")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x=cars.index, y=cars['HP'], label='HP', color='b')
sns.scatterplot(x=cars.index, y=cars['MPG'], label='MPG', color='r')

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Scatter Plot of HP vs MPG")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.scatterplot(x=cars.index, y=cars['HP'], label='HP', color='b')
sns.scatterplot(x=cars.index, y=cars['MPG'], label='MPG', color='r')
sns.scatterplot(x=cars.index, y=cars['WT'], label='WT', color='g')
sns.scatterplot(x=cars.index, y=cars['SP'], label='SP', color='m')

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Scatter Plot of HP, MPG, WT, and SP")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.scatterplot(x=cars.index, y=cars['WT'], label='WT', color='g')
sns.scatterplot(x=cars.index, y=cars['VOL'], label='VOL', color='orange')

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Scatter Plot of WT and VOL")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.scatterplot(x=cars.index, y=cars['VOL'], label='VOL', color='orange')
sns.scatterplot(x=cars.index, y=cars['SP'], label='SP', color='m')

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Scatter Plot of VOL and SP")
plt.legend()
plt.show()
```
