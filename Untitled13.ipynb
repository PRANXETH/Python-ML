```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

print(df.columns)

# Assuming the correct column name is identified as 'existing_column_name'
df = df.rename(columns={'existing_column_name': 'total_bill'})

df = df.dropna(subset=['total_bill'])

df['total_bill'] = pd.to_numeric(df['total_bill'], errors='coerce')

if df['total_bill'].empty:
    raise ValueError("'total_bill' column is empty after dropping NaN values.")

if not pd.api.types.is_numeric_dtype(df['total_bill']):
    raise TypeError("'total_bill' column is not numeric.")

sm.qqplot(df['total_bill'], line='45')
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
plt.show()
```
